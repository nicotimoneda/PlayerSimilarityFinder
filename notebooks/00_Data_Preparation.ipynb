```python
import pandas as pd
import numpy as np
import os

from numpy.ma.extras import column_stack

# Carpeta de raw y processed
raw_path = "../data/raw/"
processed_path = "../data/processed/"
os.makedirs(processed_path, exist_ok=True)

#Scrapeo de la tabla de las 5 grandes ligas europeas 2024-2025

url = "https://fbref.com/en/comps/Big5/2024-2025/stats/players/2024-2025-Big-5-European-Leagues-Stats"

tables = pd.read_html(url)
df = tables[0]
df.to_csv(os.path.join(raw_path, "big5_leagues_2024_2025_stats.csv"), index=False)
df
```
